# Laboratorio 1: Simulación de Caída Libre
En este cuaderno vamos a calcular la posición de un objeto en caída libre usando la ecuación:

$$y(t) = y_0 + v_0 t - \frac{1}{2} g t^2$$

Donde consideraremos $g = 9.8 \, m/s^2$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las constantes
g = 9.8  # m/s^2
y0 = 100 # altura inicial en metros
v0 = 0   # velocidad inicial

# Creamos un vector de tiempo de 0 a 5 segundos
t = np.linspace(0, 4.5, 100)
```

```python
# Ecuación de movimiento
y = y0 + v0*t - 0.5 * g * t**2

# Creamos el gráfico
plt.figure(figsize=(8, 5))
plt.plot(t, y, label="Trayectoria", color="blue")
plt.title("Posición vs Tiempo (Caída Libre)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.legend()
plt.show()
```
